Aplikasi ini dibuat sebagai tugas UAS untuk mencatat, mengelola, dan menampilkan data inventori barang toko kecil menggunakan bahasa pemrograman Python.

```python
# Inisialisasi data barang
data_barang = []

# === Fungsi Menu Utama ===
def tampilkan_menu():
    print("\n=== MENU UTAMA ===")
    print("1. Tambah Barang")
    print("2. Tampilkan Daftar Barang")
    print("3. Edit Barang")
    print("4. Hapus Barang")
    print("5. Cari Barang")
    print("6. Ekspor ke File CSV")
    print("7. Laporan Ringkas")
    print("8. Keluar")

# === Fungsi Tambah Barang ===
def tambah_barang():
    nama = input("Masukkan nama barang: ")
    stok = int(input("Masukkan jumlah stok: "))
    harga = int(input("Masukkan harga barang: "))
    barang = {'nama': nama, 'stok': stok, 'harga': harga}
    data_barang.append(barang)
    print(f"Barang '{nama}' berhasil ditambahkan.")

# === Fungsi Tampilkan Barang ===
def tampilkan_barang():
    if not data_barang:
        print("Belum ada data barang.")
    else:
        print("\n=== DAFTAR BARANG ===")
        for i, barang in enumerate(data_barang, start=1):
            print(f"{i}. {barang['nama']} | Stok: {barang['stok']} | Harga: Rp{barang['harga']}")

# === Fungsi Edit Barang ===
def edit_barang():
    tampilkan_barang()
    indeks = int(input("Masukkan nomor barang yang ingin diedit: ")) - 1
    if 0 <= indeks < len(data_barang):
        barang = data_barang[indeks]
        barang['nama'] = input(f"Nama baru ({barang['nama']}): ") or barang['nama']
        barang['stok'] = int(input(f"Stok baru ({barang['stok']}): ") or barang['stok'])
        barang['harga'] = int(input(f"Harga baru ({barang['harga']}): ") or barang['harga'])
        print("Barang berhasil diubah.")
    else:
        print("Nomor barang tidak valid.")

# === Fungsi Hapus Barang ===
def hapus_barang():
    tampilkan_barang()
    indeks = int(input("Masukkan nomor barang yang ingin dihapus: ")) - 1
    if 0 <= indeks < len(data_barang):
        barang = data_barang.pop(indeks)
        print(f"Barang '{barang['nama']}' berhasil dihapus.")
    else:
        print("Nomor barang tidak valid.")

# === Tambahan 1: Cari Barang ===
def cari_barang():
    keyword = input("Masukkan nama barang yang dicari: ").lower()
    hasil = [b for b in data_barang if keyword in b['nama'].lower()]
    if hasil:
        print("\n=== HASIL PENCARIAN ===")
        for i, barang in enumerate(hasil, start=1):
            print(f"{i}. {barang['nama']} | Stok: {barang['stok']} | Harga: Rp{barang['harga']}")
    else:
        print("Barang tidak ditemukan.")

# === Tambahan 2: Ekspor ke File CSV ===
import csv

def ekspor_ke_file():
    with open('data_inventory.csv', mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Nama Barang', 'Stok', 'Harga'])
        for barang in data_barang:
            writer.writerow([barang['nama'], barang['stok'], barang['harga']])
    print("Data berhasil diekspor ke 'data_inventory.csv'.")

# === Tambahan 3: Laporan Ringkas ===
def laporan_ringkas():
    total_barang = len(data_barang)
    total_stok = sum(b['stok'] for b in data_barang)
    total_nilai = sum(b['stok'] * b['harga'] for b in data_barang)

    print("\n=== LAPORAN RINGKAS ===")
    print(f"Total Jenis Barang: {total_barang}")
    print(f"Total Stok Barang: {total_stok}")
    print(f"Total Nilai Inventory: Rp{total_nilai}")

# === Fungsi Keluar ===
def keluar():
    print("Terima kasih telah menggunakan Aplikasi Inventori. Sampai jumpa!")

# === Program Utama ===
while True:
    tampilkan_menu()
    pilihan = input("Pilih menu (1-8): ")

    if pilihan == "1":
        tambah_barang()
    elif pilihan == "2":
        tampilkan_barang()
    elif pilihan == "3":
        edit_barang()
    elif pilihan == "4":
        hapus_barang()
    elif pilihan == "5":
        cari_barang()
    elif pilihan == "6":
        ekspor_ke_file()
    elif pilihan == "7":
        laporan_ringkas()
    elif pilihan == "8":
        keluar()
        break
    else:
        print("Pilihan tidak valid. Silakan pilih angka 1-8.")
```
